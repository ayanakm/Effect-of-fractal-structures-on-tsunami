```python
%%file hello.c

#include <stdio.h>
#include <math.h>

// 点の座標を表す構造体
typedef struct {
    double x;
    double y;
} Point;

// 2点間の距離を計算する関数
double distance(Point p1, Point p2) {
    double dx = p2.x - p1.x;
    double dy = p2.y - p1.y;
    return sqrt((dx * dx + dy * dy)/1.0);
}

// コッホ曲線を描く再帰関数
void drawKochCurve(Point p1, Point p2, int depth) {
    if (depth == 0) {
        printf("%.1f %.1f\n", p1.x, p1.y);
        return;
    }

    // 3等分点を計算
    Point p3 = { (2 * p1.x + p2.x) / 3, (2 * p1.y + p2.y) / 3 };
    Point p4 = { (p1.x + p2.x) / 2 - sqrt(3) * (p2.y - p1.y) / 6, (p1.y + p2.y) / 2 + sqrt(3) * (p2.x - p1.x) / 6 };
    Point p5 = { (p1.x + 2 * p2.x) / 3, (p1.y + 2 * p2.y) / 3 };

    // 再帰呼び出し
    drawKochCurve(p1, p3, depth - 1);
    drawKochCurve(p3, p4, depth - 1);
    drawKochCurve(p4, p5, depth - 1);
    drawKochCurve(p5, p2, depth - 1);
}

int main() {
    // コッホ曲線を描く始点と終点の座標
    Point start = { 100, 300 };
    Point end = { 500, 300 };

    // 再帰の深さ
    int depth = 4;

    // コッホ曲線を描く
    drawKochCurve(start, end, depth);

    return 0;
}
```

```python
import matplotlib.pyplot as plt

listx = []
listy = []

input_list = input().split()
f = len(input_list) / 2

for i in range(int(f) - 1):
    x = float(input_list[2 * i])  # x座標
    y = float(input_list[2 * i + 1])  # y座標

    listx.append(x)
    listy.append(y)

p = listx
q = listy

plt.plot(p, q)  # 点列(x,y)を黒線で繋いだプロット

# アスペクト比を設定して縦横比を等しくする
plt.axis('equal')

plt.show()  # プロットを表示
```
